```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey= pd.read_csv("/content/drive/MyDrive/ift 301/data/honey.csv")
honey.head(12)
```

```python

sns.lineplot(x='year', y='totalprod', hue='state', data=honey)
plt.title('Total Honey Production by State and Year')
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = honey[['totalprod', 'yieldpercol', 'stocks']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
avg_yield_by_state = honey.groupby('state')['yieldpercol'].mean().sort_values()
avg_yield_by_state.plot(kind='bar', figsize=(10, 6))
plt.title('Average Yield per Colony by State')
plt.xlabel('State')
plt.ylabel('Average Yield per Colony')
plt.show()
```

```python

total_production_by_year = honey.groupby('year')['totalprod'].sum()
total_stocks_by_year = honey.groupby('year')['stocks'].sum()


plt.figure(figsize=(12, 6))
plt.plot(total_production_by_year, label='Total Production')
plt.plot(total_stocks_by_year, label='Total Stocks')
plt.title('Total Production and Stocks of Honey by Year')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

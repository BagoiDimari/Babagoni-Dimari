```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
snowgoose_birds = birds[birds ['Name'] == 'Snow goose']
```

```python
Order_sum = snowgoose_birds.groupby('Order')['MinBodyMass'].sum()
plt.figure(figsize=(10, 6))
Order_sum.plot(kind='bar', color='blue', width=0.03)  
plt.title('Total minimum body mass of the "Snow Goose" bird species across different Order')
plt.xlabel('Order')
plt.ylabel('Sum of MinBodyMass')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MaxBodyMass', y='MaxWingspan', data=snowgoose_birds)
plt.title('Relationship between Maximum Body Mass and Maximum Wingspan in Snow Goose')
plt.xlabel('Max Body Mass')
plt.ylabel('Max Wingspan')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(snowgoose_birds['Category'], snowgoose_birds['MaxLength'], label='Max Length', marker='o')
plt.plot(snowgoose_birds['Category'], snowgoose_birds['MinLength'], label='Min Length', marker='o')

plt.title('Snow goose Body Lenght Over Different Categories')
plt.xlabel('Category')
plt.ylabel('Body Lenght (M)')
plt.legend()
plt.grid(True)
plt.show()
```
